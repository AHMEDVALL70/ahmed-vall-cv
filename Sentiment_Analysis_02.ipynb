```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report

# 1. تحميل البيانات (مثال تصنيف زهور Iris)
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# 2. تدريب نموذج التصنيف
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# ── 3. الرسم البياني الاحترافي (روتين أحمد فال) ──
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('AI Classification System — Ahmed Vall Jemal Dine Sidina',
             fontsize=14, fontweight='bold')

# الرسم الأول: توزيع الأصناف في البيانات (Pie Chart)
labels = iris.target_names
sizes = [sum(y_train == i) for i in range(len(labels))]
colors = ['#3498db', '#9b59b6', '#2ecc71']
axes[0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, shadow=True)
axes[0].set_title('Training Data Distribution (Classes)')

# الرسم الثاني: مصفوفة الحيرة (Confusion Matrix) لتوضيح الدقة
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[1],
            xticklabels=labels, yticklabels=labels)
axes[1].set_title('Model Performance (Confusion Matrix)')
axes[1].set_xlabel('Predicted Label')
axes[1].set_ylabel('True Label')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('classification_project_results.png', dpi=150)
plt.show()

print("📊 Classification Report:")
print(classification_report(y_test, y_pred, target_names=labels))
print("✅ تم حفظ نتائج مشروع التصنيف بنجاح!")
```
