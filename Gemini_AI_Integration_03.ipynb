```python
# ── Step 2: Setup & Configure API ───────────────────────────────
from google import genai
from google.genai import types

# Add your API key here (get one free from: https://aistudio.google.com)
GEMINI_API_KEY = 'YOUR_API_KEY_HERE'

client = genai.Client(api_key=GEMINI_API_KEY)
MODEL  = 'gemini-1.5-flash'

print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')
print('🤖 Gemini AI Client Configured!')
print(f'📌 Model: {MODEL}')
print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')
```

```python
# ── Step 3: Text Generation ──────────────────────────────────────
def generate_text(prompt):
    try:
        response = client.models.generate_content(
            model=MODEL,
            contents=prompt
        )
        return response.text
    except Exception as e:
        return f'⚠️ Error: {e}'

prompt = 'Explain the importance of Machine Learning in 3 bullet points.'
print('📝 Prompt:', prompt)
print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')
print('🤖 Gemini Response:')
print(generate_text(prompt))
```

```python
# ── Step 4: Text Summarization ───────────────────────────────────
long_text = """
Artificial Intelligence (AI) is transforming every sector of the global economy.
From healthcare diagnostics to financial fraud detection, AI systems are being
deployed at unprecedented scale. Machine learning, a subset of AI, enables
computers to learn from data without being explicitly programmed. Deep learning,
a further subset using neural networks, has achieved human-level performance in
image recognition, speech processing, and natural language understanding.
"""

summary_prompt = f'Summarize this text in 2 sentences:\n{long_text}'
print('📄 Original Text: (long paragraph)')
print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')
print('📝 Summary:')
print(generate_text(summary_prompt))
```

```python
# ── Step 5: Multi-language Support ───────────────────────────────
tasks = [
    ('🇬🇧 English', 'What is Artificial Intelligence? (1 sentence)'),
    ('🇫🇷 French',  'Qu\'est-ce que l\'Intelligence Artificielle? (1 phrase)'),
    ('🇲🇷 Arabic',  'ما هو الذكاء الاصطناعي؟ (جملة واحدة)'),
]

print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')
print('🌍 Multi-language AI Responses')
print('━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━')
for lang, prompt in tasks:
    print(f'\n{lang}:')
    print(generate_text(prompt))

print('\n✅ Project 03 — Gemini AI Integration — Completed Successfully!')
```
