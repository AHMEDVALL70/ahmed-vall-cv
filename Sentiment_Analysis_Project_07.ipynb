```python
import matplotlib.pyplot as plt
from textblob import TextBlob

# 1. قائمة التعليقات
reviews = [
    "The product is amazing and fast!",
    "I am very disappointed with the service.",
    "It is okay, not the best but works.",
    "Excellent quality, I will buy again.",
    "Terrible experience, it was a waste of money.",
    "Fast delivery and very friendly staff."
]

# 2. تحليل المشاعر وتخزين النتائج
positive_count = 0
negative_count = 0
neutral_count = 0

for review in reviews:
    polarity = TextBlob(review).sentiment.polarity
    if polarity > 0:
        positive_count += 1
    elif polarity < 0:
        negative_count += 1
    else:
        neutral_count += 1

# 3. رسم البياني (Visualization)
labels = ['Positive 😊', 'Negative 😡', 'Neutral 😐']
counts = [positive_count, negative_count, neutral_count]
colors = ['#2ecc71', '#e74c3c', '#95a5a6'] # أخضر، أحمر، رمادي

plt.figure(figsize=(8, 6))
plt.bar(labels, counts, color=colors)
plt.title('Customer Sentiment Analysis - Project 07', fontsize=14)
plt.ylabel('Number of Reviews')
plt.show()

print(f"Total Reviews: {len(reviews)}")
print(f"Positive: {positive_count}, Negative: {negative_count}, Neutral: {neutral_count}")
```
